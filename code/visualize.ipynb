```python
#Importing necessary libraries 

import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

#Importing dataset 

dt = pd.read_csv('/Users/MuskaanSandhu/09-second-exam-muskaansandhu/csv/analyze_visits_ms_data.csv')
```

#### Question #1: Walking speed analysis 

- Scatter plot of age vs walking speed with regression line
- Box plots by education level
- Line plot showing education age interaction

```python
#Scatter plot 

plt.figure(figsize=(10, 10))
sns.lmplot(x="age", y="walking_speed", data=dt, aspect=1.5, scatter_kws={"s": 20, "alpha": 0.2}, line_kws={"color": "red"})
plt.title("Age vs Walking Speed")
plt.xlabel("Age (years)")
plt.ylabel("Walking Speed (feet/second)")
plt.show()
```

```python
#Box plot 

plt.figure(figsize=(10, 10))
sns.boxplot(x="education_level", y="walking_speed", data=dt, palette="pastel")
plt.title("Walking Speed by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Walking Speed (feet/second)")
plt.show()
```

```python
#Line plot 

plt.figure(figsize=(10, 10))
sns.lineplot(x="age", y="walking_speed", hue="education_level", data=dt, marker="o", palette='pastel', alpha=0.5)
plt.title("Line Plot of Walking Speed by Education and Age")
plt.xlabel("Age (years)")
plt.ylabel("Walking Speed (feet/second)")
plt.legend(title="Education Level")
plt.show()
```

#### Question 2: Cost Analysis 

- Bar plot of mean costs by insurance type
- Box plots showing cost distributions
- Add error bars or confidence intervals

```python
#Bar plot 

plt.figure(figsize=(10, 10))
sns.barplot(x="insurance_type", y="visit_cost", data=dt, errorbar="sd", palette="pastel")
plt.title("Mean Costs by Insurance Type")
plt.xlabel("Insurance Type")
plt.ylabel("Mean Cost")
plt.show()
```

```python
#Boxplot 

plt.figure(figsize=(10, 10))
sns.boxplot(x="insurance_type", y="visit_cost", data=dt, palette="pastel")
plt.title("Cost Distribution by Insurance Type")
plt.xlabel("Insurance Type")
plt.ylabel("Cost")
plt.show()
```

#### Combined Visualizations 

- Pair plot of key variables
- Faceted plots by education/insurance
- Time trends where relevant



```python
#Pair plot 
sns.pairplot(dt[['age', 'walking_speed', 'visit_cost', 'insurance_type','education_level']], hue="education_level", palette="pastel", height=2.5)
plt.suptitle("Pair Plot of Age, Walking Speed (feet/second), Insurance Type, and Visit Cost by Education Level", y=1.1)
plt.show()
```

```python
#Faceted plot 

grid = sns.FacetGrid(dt, col="education_level", row="insurance_type", margin_titles=True)
grid.map(sns.scatterplot, "age", "walking_speed", alpha=.5, data=dt, palette="pastel", hue="insurance_type")
grid.set_axis_labels("Age (years)", "Walking Speed (feet/second)")
grid.set_titles(col_template="Education Level: {col_name}", row_template="Insurance: {row_name}")
grid.add_legend(title="Insurance Type")
grid.fig.suptitle("Faceted plots of Walking Speed and Age by Education and Insurance Type ", y=1.1)
plt.show()
```
